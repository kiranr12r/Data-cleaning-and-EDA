```python
eda_columns = [
    'OccurrenceDate',
    'ReportDate',
    'Incident Category',
    'SeverityOfIncident',
    'Location',
    'Port',
    'Defect Status',
    'Possibility Recurrence',
    'DirectCause',
    'Cause Analysis',
    'AreaOfConcern'
]

eda_data = data[eda_columns]
```

```python
eda_data['OccurrenceDate'] = pd.to_datetime(eda_data['OccurrenceDate'])
eda_data['ReportDate'] = pd.to_datetime(eda_data['ReportDate'])
```

```python
eda_data['Year'] = eda_data['OccurrenceDate'].dt.year
eda_data['Month'] = eda_data['OccurrenceDate'].dt.month
eda_data['Day'] = eda_data['OccurrenceDate'].dt.day
```

```python
eda_data['ReportDelay'] = (
    eda_data['ReportDate'] - eda_data['OccurrenceDate']
).dt.days
```

```python
eda_data['ReportDelay'].describe()
```

```python
(eda_data['ReportDelay'] < 0).sum()
```

```python
eda_data.sort_values('ReportDelay', ascending=False).head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Remove negative delays
eda_data = eda_data[eda_data['ReportDelay'] >= 0]

# Remove extreme outliers (example: >365 days)
eda_data = eda_data[eda_data['ReportDelay'] <= 365]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(eda_data['ReportDelay'], bins=40, kde=True)
plt.title("Report Delay Distribution")
plt.xlabel("Delay (Days)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=eda_data['ReportDelay'])
plt.title("Report Delay Boxplot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.boxplot(
    x='SeverityOfIncident',
    y='ReportDelay',
    data=eda_data
)

plt.xticks(rotation=45)
plt.title("Severity vs Report Delay")
plt.show()
```
